```python
class Hero:
    
    hero_health = 40
    
    def __init__(self):
        self.hero_health = Hero.hero_health
    
    def set_hero_health(self, new_hero_health):
        self.hero_health = new_hero_health
        
    def get_hero_health(self):
        return self.hero_health
```

```python
class Orc:
    
    orc_health = 7
    
    def __init__(self):
        self.orc_health = Orc.orc_health
    
    def set_orc_health(self, new_orc_health):
        self.orc_health = new_orc_health
        
    def get_orc_health(self):
        return self.orc_health
```

```python
class Dragon:
    
    dragon_health = 20
    
    def __init__(self):
        self.dragon_health = Dragon.dragon_health
    
    def set_dragon_health(self, new_dragon_health):
        self.dragon_health = new_dragon_health
        
    def get_dragon_health(self):
        return self.dragon_health
```
